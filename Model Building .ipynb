## ENCODING QUESTIONS:

```python
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk import ngrams

 
encode_que = data['Question']
 
# Create a Vectorizer Object
tfidfvectorizer = TfidfVectorizer(analyzer='word',stop_words= 'english')
 
que_matrix = tfidfvectorizer.fit_transform(encode_que.values.astype('U'))

que_tokens = tfidfvectorizer.get_feature_names_out()

df_que_tfvect = pd.DataFrame(data = que_matrix.toarray(),columns = que_tokens)
df_que_tfvect
```

```python
# Summarizing the Encoded Answer
print("Encoded Answer Shape")
print("-------------------")
print(que_matrix.shape)
```

## ENCODING ANSWER:

```python
encode_ans = data['multiple_choice_answer']
 
ans_matrix = tfidfvectorizer.fit_transform(encode_ans.values.astype('U'))

ans_tokens = tfidfvectorizer.get_feature_names_out()

df_ans_tfvect = pd.DataFrame(data = ans_matrix.toarray(),columns = ans_tokens)
df_ans_tfvect
```
